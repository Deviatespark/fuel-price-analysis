```python
import pandas as pd
import matplotlib.pyplot as plt


def cargar_datos(archivo_csv):
    """Carga el archivo CSV y devuelve un DataFrame."""
    try:
        df = pd.read_csv(archivo_csv, dtype=str) 
        return df
    except FileNotFoundError:
        print("Error: No se encontró el archivo '{archivo_csv}'.")
        return None


def filtrar_por_provincia(df, provincia_interes):
    """Filtra el DataFrame por una provincia específica."""
    if "Provincia" not in df.columns:
        print("Error: La columna 'Provincia' no existe en el dataset.")
        return None

    df_provincia = df[df["Provincia"].str.upper() == provincia_interes].copy()
    
    if df_provincia.empty:
        print("No hay datos para la provincia {provincia_interes}.")
    
    return df_provincia


def limpiar_precio(valor):
    """Convierte precios de '1,529' a 1.529 (float)."""
    if isinstance(valor, str):
        return float(valor.replace(",", ".")) if valor.replace(",", "").replace(".", "").isdigit() else None
    return None


def limpiar_datos(df):
    """Convierte las columnas de precios a valores numéricos y elimina NaN."""
    if df is None:
        return None

    columnas_precios = ["Precio gasolina 95 E5", "Precio gasóleo A"]
    
    for col in columnas_precios:
        if col in df.columns:
            df[col] = df[col].apply(limpiar_precio)
        else:
            print(" Advertencia: La columna '{col}' no existe en el dataset.")
    
    df.dropna(subset=columnas_precios, inplace=True)
    return df


def graficar_precios(df, provincia_interes):
    
    if df is None or df.empty:
        print("No hay datos suficientes para graficar.")
        return

    plt.figure(figsize=(10, 5))

    if "Precio gasolina 95 E5" in df.columns:
        plt.hist(df["Precio gasolina 95 E5"], bins=10, alpha=0.5, label="Gasolina 95 E5", color="red")

    if "Precio gasóleo A" in df.columns:
        plt.hist(df["Precio gasóleo A"], bins=10, alpha=0.5, label="Gasóleo A", color="blue")

    plt.xlabel("Precio (€)")
    plt.ylabel("Número de estaciones")
    plt.title(f"Distribución de precios en {provincia_interes}")
    plt.legend()
    plt.show()


def main():
    
    archivo_csv = "gii32_act1_precios_carburantes_24.csv"
    provincia_interes = "MADRID"

    # Cargar datos
    df = cargar_datos(archivo_csv)
    if df is None:
        return

    # Filtrar por provincia
    df_provincia = filtrar_por_provincia(df, provincia_interes)
    if df_provincia is None or df_provincia.empty:
        return

    # Limpiar datos
    df_provincia = limpiar_datos(df_provincia)

    # Graficar precios
    graficar_precios(df_provincia, provincia_interes)


# Ejecutar el script
if __name__ == "__main__":
    main()
```
